```python
from sklearn.linear_model import Lasso

# Create the Lasso regression model
lasso = Lasso(alpha=0.1)

# Fit the model to the training data
lasso.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = lasso.predict(X_test)

# Evaluate the model
score = lasso.score(X_test, y_test)
print(f"The model's R^2 score is: {score:.2f}")
mse = mean_squared_error(y_test, y_pred)
print("Mean squared error:", mse)
```

```python
from sklearn.linear_model import Ridge

# Create the Ridge regression model
ridge = Ridge(alpha=0.1)

# Fit the model to the training data
ridge.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = ridge.predict(X_test)

# Evaluate the model
r_score = ridge.score(X_test, y_test)
print(f"The model's R^2 score is: {r_score:.2f}")
r_mse = mean_squared_error(y_test, y_pred)
print("Mean squared error:", r_mse)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Transform the features into polynomial features
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Create the polynomial regression model
poly_reg = LinearRegression()

# Fit the model to the training data
poly_reg.fit(X_train_poly, y_train)

# Make predictions on the testing data
y_pred = poly_reg.predict(X_test_poly)

# Evaluate the model
poly_reg_r2 = poly_reg.score(X_test_poly, y_test)
print(f"The model's R^2 score is: {poly_reg_r2:.2f}")
poly_reg_mse = mean_squared_error(y_test, y_pred)
print("Mean squared error:", poly_reg_mse)
```

```python
from sklearn.ensemble import RandomForestRegressor
# Create the Random Forest regression model
rf = RandomForestRegressor(n_estimators=100, random_state=0)

# Fit the model to the training data
rf.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = rf.predict(X_test)

# Evaluate the model
rf_r2 = rf.score(X_test, y_test)
print(f"The model's R^2 score is: {rf_r2:.2f}")
rf_mse = mean_squared_error(y_test, y_pred)
print("Mean squared error:", rf_mse)
```

#### Note that lower values of MSE indicate better performance, while higher values of R² indicate better performance.Therefore, depending on your goals and priorities, you may want to sort the table by either MSE or R².

we got lower value of mse in polynomial regression so we'll apply some hypertuning parameters in order to improve this.


```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.pipeline import Pipeline

```

```python
# Define the hyperparameters to tune
param_grid = {
    'polynomialfeatures__degree': [1, 2, 3, 4],
    'linearregression__fit_intercept': [True, False]
}

# Create the polynomial regression model
poly_reg = PolynomialFeatures(include_bias=False)
lin_reg = LinearRegression()

# Create a pipeline with PolynomialFeatures and LinearRegression
pipeline = Pipeline([
    ('polynomialfeatures', poly_reg),
    ('linearregression', lin_reg)
])

# Perform a grid search over the hyperparameters
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train_poly, y_train)

# Print the best hyperparameters and the corresponding MSE
print("Best hyperparameters: ", grid_search.best_params_)
print("Best mean squared error: ", -grid_search.best_score_)

# Evaluate the final model on the test set
y_pred = grid_search.predict(X_test_poly)
print("Test mean squared error: ", mean_squared_error(y_test, y_pred))
```
